```python
# Iterate through the pairs and populate the dictionary
for tree_leaf in tree_leafs:
  feat_dict = Morphology.feats_to_dict(tree_leaf)
  pos = spacy.explain(feat_dict["POS"])
  for attribute, value in feat_dict.items():

    # Ensure that the outer dictionary for POS is initialized
    if not pos in poss_vals_for_atts_per_pos:
      poss_vals_for_atts_per_pos[pos] = {}

    # Ensure that the set for the attribute is initialized in the global dictionary
    if not attribute in poss_vals_for_atts:
      poss_vals_for_atts[attribute] = set()

    # Ensure that the inner dictionary for each attribute under POS is initialized
    if not attribute in poss_vals_for_atts_per_pos[pos]:
      poss_vals_for_atts_per_pos[pos][attribute] = set()

    # Add the value to the global set for the attribute
    poss_vals_for_atts[attribute].add(value)

    # Add the value to the set for the attribute under the specific POS
    poss_vals_for_atts_per_pos[pos][attribute].add(value)
```
